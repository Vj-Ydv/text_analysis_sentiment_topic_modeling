## **CSV to Text conversion**

```python
%cd /content/drive/My Drive/Colab Notebooks/timesofindia_csv
```

```python
#convertig .csv file to .txt file with each column separated by a TAB

import csv
import os

cwd = os.getcwd()
#print(cwd)

for file in os.listdir('.'):  # use the directory name here

    file_name, file_ext = os.path.splitext(file)

    print(file_name)

    if file_ext == '.csv':
        with open(file,'r') as csv_file:
            csv_reader = csv.reader(csv_file)

            next(csv_reader)  ## skip one line (the first one)

            newfile = file + '.txt'

            for line in csv_reader:
                with open(newfile, 'a') as new_txt:    #new file has .txt extn
                    txt_writer = csv.writer(new_txt, delimiter = '\t') #writefile
                    txt_writer.writerow(line)   #write the lines to file`
```

```python
# Python program to count the 
# number of lines in a original text file
  
  
# Opening a file
file = open("toi_complete.txt","r")
Counter = 0
  
# Reading from file
Content = file.read()
CoList = Content.split("\n")
  
for i in CoList:
    if i:
        Counter += 1
          
print("This is the number of lines in the file")
print(Counter)
```

## **Removing those lines which do not have 3 columns maintained as others**

```python
f = open("toi_complete.txt","r", encoding="utf8", errors='replace').read()

with open("toi_full.txt", "w") as output: 
        for line in f.split('\n'):
          columns = line.split('\t')
          columns = [col.strip() for col in columns]
          col_size= len(columns)

          if col_size==3:
                output.write(line + '\n')
                
```
